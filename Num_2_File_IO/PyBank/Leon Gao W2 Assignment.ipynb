```python
# Import the pathlib and csv liabrary
from pathlib import Path
import csv

# Set the file input path
csvpath = Path("Resources/budget_data.csv")

# Initialize metric variables, list and dictionary
total = 0
count = 0
daily_pnl = []
dict_changes = {}
largest_gain = 0
largest_loss = 0

# Open the input path a file object
with open(csvpath,"r") as csvfile:
    #Pass in the csv file to the csv.reader() function and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter = ",")
    
    # Go to the next row from the start of the file
    header = next(csvreader)
    
    # Read each row after the header
    for row in csvreader:
        
        # Sum the total and count variables
        total += int(row[1])
        count += 1
        
        # Update daily P&L list
        daily_pnl.append(row[1])
        # Update daily changes dictionary
        if count > 1:
            dict_changes[row[0]] = int(row[1]) - int(daily_pnl[count-2])
     
    # Iterate the dictionary to determine greatest increase and decrease in profits and their associated date
    for key, value in dict_change.items():
        
        if int(value) > largest_gain:
            largest_gain = int(value)
            largest_gain_date = key
        
        elif int(value) < largest_loss:
            largest_loss = int(value)
            largest_loss_date = key        
  
    # Calculate the average change over the period and round to the nearest 2 decimal places
    average = round((int(daily_pnl[-1])-int(daily_pnl[0]))/(len(daily_pnl)-1),2)

# Set the path for output.txt
output_path = Path("output.txt")
# Open the output path as file object
with open(output_path,"w") as file:
    
    # Write Financial Analysis to the file
    file.write("Financial Analysis")
    file.write("\n-------------------------------")
    file.write(f"\nTotal Months: {count}")
    file.write(f"\nTotal: ${total}")
    file.write(f"\nAverage Change: ${average}")
    file.write(f"\nGreatest Increase in Profits: {largest_gain_date} (${largest_gain})")
    file.write(f"\nGreatest Decrease in Profits: {largest_loss_date} (${largest_loss})")
```
